```python
# Import necessary libraries
import pandas as pd
import pycountry_convert as pc
import plotly.express as px
import plotly.io as pio
```

```python
# Read the life expectancy and GDP per capita datasets
expectancy = pd.read_excel("life_expentancy.xlsx")
gdp = pd.read_excel("gdp_per_capita.xlsx")
```

```python
# Reshape the GDP per capita dataset using the melt function
gdp = pd.melt(gdp, id_vars='name',var_name='Year', value_name='gdp')
gdp.head()
```

```python
# Reshape the life expectancy dataset using the melt function
expectancy = pd.melt(expectancy, id_vars='name',var_name='Year', value_name='LifeExpectancy')
expectancy.head()
```

```python
# Check the data types of the columns again
gdp.dtypes
```

```python
expectancy.dtypes
```

```python
# Display the last few rows of the reshaped GDP per capita dataset
gdp.tail()
```

```python
# Merge the GDP per capita and life expectancy datasets based on common columns
gdp_exp = pd.merge(gdp, expectancy)
gdp_exp.head()
```

```python
# Read the population dataset
population = pd.read_excel("population.xlsx")
population.head()
```

```python
# Reshape the population dataset using the melt function
population = pd.melt(population, id_vars='name',var_name='Year', value_name='population')
population.head()
```

```python
# Check the data types of the columns in the population dataset
population.dtypes
```

```python
# Check the data types of the columns in the merged GDP and life expectancy dataset
gdp_exp.dtypes
```

```python
# Merge the GDP, life expectancy, and population datasets based on common columns
gdp_exp_pop = pd.merge(gdp_exp, population)
gdp_exp_pop.head()
```

```python
# Check the data types of the columns in the merged dataset
gdp_exp_pop.dtypes
```

```python
# Convert the "Year" and "population" columns to integer data type
gdp_exp_pop["Year"] = gdp_exp_pop["Year"].astype('int64')
gdp_exp_pop["population"] = gdp_exp_pop["population"].astype('int64')
gdp_exp_pop.dtypes
```

```python
# Replace country names with correct spellings and formatting
gdp_exp_pop["name"] = gdp_exp_pop["name"].replace({"Cote d'Ivoire": "Côte d'Ivoire","Congo, Dem. Rep.": "Democratic Republic of the Congo","Congo, Rep.":"Republic of the Congo", "Micronesia, Fed. Sts.":"Federated States of Micronesia", "Hong Kong, China": "Hong Kong","Lao": "Lao People's Democratic Republic", "St. Vincent and the Grenadines":"Saint Vincent and the Grenadines", "Macedonia, FYR":"Macedonia"})
```

```python
# Define a lambda function to convert country names to country codes
f = lambda x: pc.country_name_to_country_alpha2(x, cn_name_format="default")
gdp_exp_pop['CountryCode'] = gdp_exp_pop['name'].apply(f)
```

```python
gdp_exp_pop.head()
```

```python
# Defining a function to convert country names to continent codes
def convert(row):
    country_code = pc.country_name_to_country_alpha2(row.country, cn_name_format = "default")
    conti_code = pc.country_alpha2_to_continent_code(country_code)
    return conti_code
```

```python
# Applying the 'convert' function to create a new column 'continent' in 'gdp_exp_pop' dataframe
gdp_exp_pop["coninent"] = gdp_exp_pop.apply(convert, axis=1)
gdp_exp_pop.head()
```

```python
# Reading the 'countryContinent.xlsx' file into a pandas DataFrame 'cont'
cont = pd.read_excel("countryContinent.xlsx")
cont.head()
```

```python
# Filtering rows in 'cont' DataFrame where 'code_2' is equal to "TL" (example: Timor-Leste)
timor = cont[cont["code_2"]== "TL"]
timor.head()
```

```python
# Displaying information about 'cont' DataFrame
cont.info()
```

```python
# Selecting specific columns 'name' and 'continent' from 'cont' DataFrame
cont = cont[["name","continent"]]
cont.head()
```

```python
# Merging 'gdp_exp_pop' and 'cont' DataFrames based on 'name' column and storing the result in 'merged_df'
merged_df = pd.merge(gdp_exp_pop, cont, how = 'left', on = "name")
```

```python
merged_df.shape
```

```python
# Checking the shape of 'gdp_exp_pop' DataFrame
gdp_exp_pop.shape
```

```python
# Checking the data types of 'merged_df' DataFrame
merged_df.dtypes
```

```python
# Filtering rows in 'merged_df' DataFrame where 'Year' is between 1900 and 2023 (inclusive)
merged_df = merged_df[(merged_df["Year"]>= 1900) & (merged_df["Year"]<= 2023)]
merged_df.head()
```

```python
# Resetting the index of 'merged_df' DataFrame and dropping the 'index' column
merged_df = merged_df.reset_index()
merged_df = merged_df.drop(["index"], axis=1)
merged_df.head()
```

```python
# Creating a copy of 'merged_df' DataFrame and assigning it to 'df'
df = merged_df.copy()
df.head()
```

```python
df.tail()
```

```python
# Creating an animated scatter plot using Plotly Express
import plotly.io as pio
fig = px.scatter(df,x="gdp", 
                 y="LifeExpectancy",color="continent", 
                 animation_frame= "Year",animation_group="name",
                 labels={"LifeExpectancy":"Life Expectancy", "gdp":"GDP per Capita", "continent":"Continent"},
                 size="population",
                 hover_name="name", 
                 log_x=True, size_max=100, 
                 range_x=[200, 60000], 
                 range_y=[0,100],
                 color_discrete_sequence=px.colors.qualitative.G10).update_layout(
                          font_family="Georgia, serif",
                          font_color="black",
                          legend_title_font_color="green", paper_bgcolor='rgba(0,0,0,0)',
                          plot_bgcolor='rgba(0,0,0,0)').update_xaxes(zeroline=False, showgrid=False,
                                                                     title_font_family="Georgia, serif").update_yaxes(
                          zeroline=False, showgrid=False)
fig.update_layout(title_text="<b>Exploring GDP, LifeExpectancy and Population (1900-2023)<b>", title_x=0.5,
                  title_font_size = 16,
                        font_family="Georgia, serif", title_font_family="Georgia, serif", 
                        title_font_color="black", legend_title_font_color="green",
                        yaxis = dict(tickfont = dict(size=14)),
                        xaxis = dict(tickfont = dict(size=20))
                        ,sliders = [dict(font={'size': 17})] 
                 )
fig.show(pio="browser")
```
